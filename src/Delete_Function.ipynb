### Define a function that deletes publishers

```python
def delete_publisher(id):
    query = text("""
        DELETE FROM publishers
        WHERE publisher_id = :id
    """)

    with engine.begin() as conn:
        conn.execute(query, {"id": id})

    print("Publisher deleted!")
```
