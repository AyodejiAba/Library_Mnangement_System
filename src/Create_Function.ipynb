### Importing the required packages

```python
from sqlalchemy import create_engine, text
import pandas as pd
```

### Creating a connection string and engine that connects to the mysql

```python
schema = "liane_library"
host = "127.0.0.1"
user = "root"
password = "Ayodeji007!"
port = 3306

connection_string = (
    f"mysql+pymysql://{user}:{password}@{host}:{port}/{schema}"
)

engine = create_engine(connection_string)
```

### define a function that create Publishers 

```python
def create_publisher(name, country):
    query = text("""
        INSERT INTO publishers (publisher_name, country)
        VALUES (:name, :country)
    """)

    with engine.begin() as conn:
        conn.execute(query, {
            "name": name,
            "country": country
        })

    print("Publisher added successfully!")

# The engine.begin initiates transfer of data from python to sql.
```

### Define a function that creates author

```python
def create_author(first_name, last_name, email, nationality):
    query = text("""
        INSERT INTO authors (first_name, last_name, email, nationality)
        VALUES (:first_name, :last_name, :email, :nationality)
    """)

    with engine.begin() as conn:
        conn.execute(query, {
            "first_name": first_name,
            "last_name": last_name,
            "email": email,
            "nationality": nationality
        })

    print("Author added successfully.")
```

### Define a function that creates book copy

```python
def create_book_copy(isbn, status):
    query = text("""
        INSERT INTO book_copies (isbn, status)
        VALUES (:isbn, :status)
    """)

    with engine.begin() as conn:
        conn.execute(query, {
            "isbn": isbn,
            "status": status
        })

    print("Book copy added.")
```

### Define a function that creates book loan 

```python
def create_book_loan(isbn, member_id, borrowed_date, due_date, returned_date=None):
    query = text("""
        INSERT INTO book_loans
        (isbn, member_id, borrowed_date, due_date, returned_date)
        VALUES
        (:isbn, :member_id, :borrowed_date, :due_date, :returned_date)
    """)

    with engine.begin() as conn:
        conn.execute(query, {
            "isbn": isbn,
            "member_id": member_id,
            "borrowed_date": borrowed_date,
            "due_date": due_date,
            "returned_date": returned_date
        })

    print("Loan created.")
```

### Define a function that creates book

```python
def create_book(isbn, title, publication_year, language, publisher_id, author_id):
    query = text("""
        INSERT INTO books
        (isbn, title, publication_year, language, publisher_id, author_id)
        VALUES
        (:isbn, :title, :publication_year, :language, :publisher_id, :author_id)
    """)

    with engine.begin() as conn:
        conn.execute(query, {
            "isbn": isbn,
            "title": title,
            "publication_year": publication_year,
            "language": language,
            "publisher_id": publisher_id,
            "author_id": author_id
        })

    print("Book created.")
```
