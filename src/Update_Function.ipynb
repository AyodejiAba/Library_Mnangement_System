### Define a function that updates publishers

```python
def update_publisher(id, new_name, new_country):
    query = text("""
        UPDATE publishers
        SET publisher_name = :name,
            country = :country
        WHERE publisher_id = :id
    """)

    with engine.begin() as conn:
        conn.execute(query, {
            "id": id,
            "name": new_name,
            "country": new_country
        })

    print("Publisher updated!")
```

### Define a function that updates authors

```python
def update_author(author_id, first_name, last_name, email, nationality):
    query = text("""
        UPDATE authors
        SET first_name = :first_name,
            last_name = :last_name,
            email = :email,
            nationality = :nationality
        WHERE author_id = :author_id
    """)

    with engine.begin() as conn:
        conn.execute(query, {
            "author_id": author_id,
            "first_name": first_name,
            "last_name": last_name,
            "email": email,
            "nationality": nationality
        })

    print("Author updated.")
```

### Define a function that updates book_copy

```python
def update_book_copy(copy_id, isbn, status):
    query = text("""
        UPDATE book_copies
        SET isbn = :isbn,
            status = :status
        WHERE copy_id = :copy_id
    """)

    with engine.begin() as conn:
        conn.execute(query, {
            "copy_id": copy_id,
            "isbn": isbn,
            "status": status
        })

    print("Book copy updated.")
```

### Define a function that updates book_loan

```python
def update_book_loan(loan_id, isbn, member_id,
                     borrowed_date, due_date, returned_date):

    query = text("""
        UPDATE book_loans
        SET isbn = :isbn,
            member_id = :member_id,
            borrowed_date = :borrowed_date,
            due_date = :due_date,
            returned_date = :returned_date
        WHERE loan_id = :loan_id
    """)

    with engine.begin() as conn:
        conn.execute(query, {
            "loan_id": loan_id,
            "isbn": isbn,
            "member_id": member_id,
            "borrowed_date": borrowed_date,
            "due_date": due_date,
            "returned_date": returned_date
        })

    print("Loan updated.")
```

### Define a function that updates books

```python
def update_book(isbn, title, publication_year, language, publisher_id, author_id):
    query = text("""
        UPDATE books
        SET
            title = :title,
            publication_year = :publication_year,
            language = :language,
            publisher_id = :publisher_id,
            author_id = :author_id
        WHERE isbn = :isbn
    """)

    with engine.begin() as conn:
        conn.execute(query, {
            "isbn": isbn,
            "title": title,
            "publication_year": publication_year,
            "language": language,
            "publisher_id": publisher_id,
            "author_id": author_id
        })

    print("Book updated.")
```

### Define a function that updates members
